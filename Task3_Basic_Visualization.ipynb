```python
# Task3 Basic Visualization
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("iris_cleaned.csv")

# Bar Plot: Species count
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=df, palette="Set2")
plt.title("Count of each Iris Species")
plt.show()

# Pie Chart: Species distribution 
species_counts = df['species'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(species_counts, labels=species_counts.index, autopct="%1.1f%%", colors=["#66b3ff","#99ff99","#ffcc99"])
plt.title("Species Distribution (Pie Chart)")
plt.show()

#  Line Plot: Sepal length by index 
plt.figure(figsize=(8,4))
plt.plot(df.index, df['sepal_length'], label="Sepal Length", color="blue")
plt.title("Line Plot of Sepal Length across Samples")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.legend()
plt.show()

print("\n Task 3 (Basic Visualization) Completed")
```
